```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('data.txt', delimiter='\s+', header = None)
x = np.array(df[0]).reshape(-1, 1)
y = np.array(df[1]).reshape(-1, 1)

model = linear_model.LinearRegression()
model.fit(x, y)

print("slope: ",model.coef_.round(5), "\nintercept: ", model.intercept_.round(5))

plt.plot(x, model.predict(x), color = 'red')
j = plt.scatter(x, y, color = 'darkblue')
```

```python
n_points = 60
slope = 1.5
intercept = -3
std_dev = 5

x = np.linspace(0, 10, n_points)
y_true = np.array(slope * x + intercept)
y = np.array(y_true + np.random.normal(0, std_dev, size=n_points))

data = np.column_stack((x, y))

np.savetxt('data2.txt', data, fmt='%.2f')

# plt.scatter(x, y, marker = 'x', label='Data Points', color = 'darkblue')
# l = plt.plot(x, y_true, color='red', label='True Line')
```
